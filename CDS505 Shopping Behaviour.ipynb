# Transforming Data

```python
#del df_new

df_new = df.copy()
df_new
```

## Binning Age

```python
# Define the name of each bin
cut_label = ['Young', 'Middle', 'Senior', 'Old']

# Define the boundaries of the 4 bins
cut_bins = [15, 30, 50, 65, 80]

# Divide the data into 4 bins with the manually-defined range
df_new['Age Category'] = pd.cut(df_new['Age'], bins = cut_bins, labels = cut_label)

# View the count of each bin
df_new['Age Category'].value_counts()
```

##

## Converting Shipping Type

```python
df_new['Shopping Type'] = df_new['Shipping Type']
df_new['Shopping Type'] = df_new['Shipping Type'].replace({'Express':'Online', 'Free Shipping':'Online', 'Next Day Air':'Online', 'Standard':'Online','2-Day Shipping':'Online','Store Pickup':'Offline'})
```

```python
df_new['Shopping Type'].value_counts()
```

```python
df_new['Shopping Type']
```

##

## Binning Review Ratings

```python
# Define the name of each bin
cut_label = ['Poor', 'Fair', 'Average', 'Good', 'Superb']

# Define the boundaries of the 5 bins
cut_bins = [0, 1, 2, 3, 4, 5]

# Divide the data into 5 bins with the manually-defined range
df_new['Rating Summary'] = pd.cut(df_new['Review Rating'], bins = cut_bins, labels = cut_label)

# View the count of each bin
df_new['Rating Summary'].value_counts()
```

##

## Binning Previous Purchases

```python
# Define the name of each bin
cut_label = ['New', 'Regular', 'Loyal']

# Define the boundaries of the 3 bins
cut_bins = [0, 10, 25, 50]

# Divide the data into 3 bins with the manually-defined range
df_new['Customer Loyalty'] = pd.cut(df_new['Previous Purchases'], bins = cut_bins, labels = cut_label)

# View the count of each bin
df_new['Customer Loyalty'].value_counts()
```
